```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
inputimage  = cv.imread("/content/IMG_20221212_214334.jpg")
inputimage = cv.resize(inputimage, (500, 500))
```

```python
inputimage2  = cv.imread("/content/drive/MyDrive/IMG_20221212_214036.jpg")
inputimage2 = cv.resize(inputimage2, (500, 500))
```

```python
inputimage.shape
```

```python
plt.figure()
plt.imshow(cv.cvtColor(inputimage, cv.COLOR_BGR2GRAY))
```

```python
from google.colab.patches import cv2_imshow
```

```python
t_lower = 100  # Lower Threshold
t_upper = 400  # Upper threshold
  
# Applying the Canny Edge filter
edge = cv.Canny(inputimage, t_lower, t_upper)
edge1 = cv.Canny(inputimage2,t_lower,t_upper)
cv2_imshow(edge)
cv2_imshow(edge1)
```

```python
# The line shouldn't lie between these y-axes . If lies Defected.
low = 180  
high = 230
```

```python
tp = inputimage.copy()
lines1 = cv.HoughLines(edge,1,np.pi/180,58)
print(len(lines1))
# rho,theta = lines1[4][0]
# a = np.cos(theta)
# b = np.sin(theta)
# x0 = a * rho
# y0 = b * rho
# x1 = int(x0 + 1000 *(-b))
# y1 = int(y0 + 1000 *(a))
# x2 = int(x0 - 1000 *(-b))
# y2 = int(y0 - 1000 * (a))
# print((y1+y2)/2)
# cv.line(tp,(x1,y1),(x2,y2),(0,0,255),1)


# rho,theta = lines1[5][0]
# a = np.cos(theta)
# b = np.sin(theta)
# x0 = a * rho
# y0 = b * rho
# x1 = int(x0 + 1000 *(-b))
# y1 = int(y0 + 1000 *(a))
# x2 = int(x0 - 1000 *(-b))
# y2 = int(y0 - 1000 * (a))
# print((y1+y2)/2)
# cv.line(tp,(x1,y1),(x2,y2),(0,255,0),1)


# rho,theta = lines1[6][0]
# a = np.cos(theta)
# b = np.sin(theta)
# x0 = a * rho
# y0 = b * rho
# x1 = int(x0 + 1000 *(-b))
# y1 = int(y0 + 1000 *(a))
# x2 = int(x0 - 1000 *(-b))
# y2 = int(y0 - 1000 * (a))
# print((y1+y2)/2)
# cv.line(tp,(x1,y1),(x2,y2),(255,0,0),1)

# rho,theta = lines1[7][0]
# a = np.cos(theta)
# b = np.sin(theta)
# x0 = a * rho
# y0 = b * rho
# x1 = int(x0 + 1000 *(-b))
# y1 = int(y0 + 1000 *(a))
# x2 = int(x0 - 1000 *(-b))
# y2 = int(y0 - 1000 * (a))
# print((y1+y2)/2)
# cv.line(tp,(x1,y1),(x2,y2),(255,255,0),1)
# # cv2_imshow(tp)

# print("end")
flag= False
for line in lines1:
  rho , theta = line[0]
  if(theta <1.58):
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a * rho
    y0 = b * rho
    x1 = int(x0 + 1000 *(-b))
    y1 = int(y0 + 1000 *(a))
    x2 = int(x0 - 1000 *(-b))
    y2 = int(y0 - 1000 * (a))
    temp= (y1+y2)/2
    if((temp > 180 and temp < 230 )):
      flag = True
    cv.line(tp,(x1,y1),(x2,y2),(0,0,255),5)

if(flag):
  print("The Bottle is Defected.")
else:
  print("The Bottle is Not Defected.")
cv2_imshow(tp)
```
